```python
# Function to detect and remove outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Outliers in {column}:")
    print(outliers)
    
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df
```

```python
# Function to print correlated columns
def print_correlated_columns(correlation_matrix, column, threshold=0.7):
    correlated_columns = correlation_matrix[column][abs(correlation_matrix[column]) > threshold].index.tolist()
    correlated_columns.remove(column)
    print(f"Columns correlated with {column}: {correlated_columns}")

# Calculate correlation matrix
correlation_matrix = data.corr()

# Iterate over columns
for column in data.columns:
    if data[column].dtype in ['int64', 'float64']:  # Check if column is numerical
        print(f"\nColumn: {column}")
        print_correlated_columns(correlation_matrix, column)
        data = remove_outliers(data, column)

print("Data shape after removing outliers:", data.shape)


```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Function to plot distribution
def plot_distribution(data, column, title):
    plt.figure(figsize=(8,6))
    plt.hist(data[column].dropna(), bins=50, density=True, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title(title)
    plt.xlabel(column)
    plt.ylabel("Density")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Function to identify and transform distribution
def identify_distribution(data, column):
    col_data = data[column].dropna()

    # Perform normality test
    stat, p = stats.normaltest(col_data)
    if p < 0.05:
        print(f"{column} does not follow a normal distribution")
        
        # Try to identify the distribution
        distributions = ['norm', 'expon', 'uniform', 'poisson']
        results = []
        for dist in distributions:
            try:
                params = getattr(stats, dist).fit(col_data)
                d, p_value = stats.kstest(col_data, dist, args=params)
                results.append((dist, p_value))
            except Exception as e:
                continue
        
        if not results:
            print(f"No valid distribution fit for {column}. Skipping transformation.")
            return col_data

        # Get the best fit distribution
        best_dist = max(results, key=lambda x: x[1])  # best = highest p-value
        print(f"Best fit distribution for {column}: {best_dist[0]} (p={best_dist[1]:.4f})")

        # Apply transformation
        try:
            if best_dist[0] == 'expon':
                transformed_data = np.log(col_data + 1e-5)
            elif best_dist[0] == 'uniform':
                scaled = (col_data - col_data.min()) / (col_data.max() - col_data.min())
                transformed_data = stats.norm.ppf(scaled.clip(1e-5, 1 - 1e-5))
            elif best_dist[0] == 'poisson':
                transformed_data = np.sqrt(col_data)
            else:
                transformed_data = col_data  # already normal
        except Exception as e:
            print(f"Transformation failed for {column}: {e}")
            transformed_data = col_data

        return transformed_data
    else:
        print(f"{column} follows a normal distribution")
        return data[column]

# Iterate over numeric columns
for column in data.select_dtypes(include=['int64', 'float64']).columns:
    print(f"\nColumn: {column}")
    plot_distribution(data, column, f"Original Distribution of {column}")
    transformed_data = identify_distribution(data, column)
    data[column] = transformed_data
    plot_distribution(data, column, f"Transformed Distribution of {column}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Function to plot distribution
def plot_distribution(data, column, title):
    plt.figure(figsize=(8,6))
    plt.hist(data[column].dropna(), bins=50, density=True, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title(title)
    plt.xlabel(column)
    plt.ylabel("Density")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Function to normalize column using the best transformation
def normalize_column(data, column):
    col_data = data[column].dropna()

    # Skip if column has constant value
    if col_data.nunique() <= 1:
        print(f"{column} has constant value. Skipping.")
        return data[column]

    transformations = {
        'none': lambda x: x,
        'log': lambda x: np.log(x + 1e-5) if np.all(x > 0) else None,
        'sqrt': lambda x: np.sqrt(x) if np.all(x >= 0) else None,
        'boxcox': lambda x: stats.boxcox(x + 1e-5)[0] if np.all(x > 0) else None,
        'yeojohnson': lambda x: stats.yeojohnson(x)[0],
    }

    best_p = -np.inf
    best_trans = 'none'
    best_data = col_data

    for name, func in transformations.items():
        try:
            transformed = func(col_data)
            if transformed is not None and np.all(np.isfinite(transformed)):
                stat, p = stats.normaltest(transformed)
                if p > best_p:
                    best_p = p
                    best_trans = name
                    best_data = transformed
        except Exception as e:
            continue

    print(f"Best transformation for {column}: {best_trans} (p={best_p:.4f})")
    return pd.Series(best_data, index=col_data.index)

# Apply transformation to each numeric column
for column in data.select_dtypes(include=['int64', 'float64']).columns:
    print(f"\nColumn: {column}")
    plot_distribution(data, column, f"Original Distribution of {column}")
    transformed_column = normalize_column(data, column)
    data[column].loc[transformed_column.index] = transformed_column
    plot_distribution(data, column, f"Transformed Distribution of {column}")
```

```python
#feature engineering
'''
1.BMI = Weight / (Height/100)
Max_HR = 220 - Age
2.Intensity = Heart_Rat4.e / Max_HR
if Intensity < 0.6: Low
elif Intensity < 0.75: Moderate
else: High
3.Cal_per_min = Calories / Duration
4.Temp_Deviation = Body_Temp - 37.0
5.WHtR = Weight / Height
6.VO2_proxy = Heart_Rate * Duration
7.HR_Temp = Heart_Rate * Body_Temp
'''
import pandas as pd
import numpy as np

# Assuming data is your original DataFrame
# Example: data = pd.read_csv("your_data.csv")

# BMI
data['BMI'] = data['Weight'] / ((data['Height'] / 100) ** 2)

# Max Heart Rate (220 - age)
data['Max_Heart_Rate'] = 220 - data['Age']

# Intensity of workout
data['Intensity'] = data['Heart_Rate'] / data['Max_Heart_Rate']

# Intensity Category
def classify_intensity(i):
    if i < 0.6:
        return 'Low'
    elif i < 0.75:
        return 'Moderate'
    else:
        return 'High'

data['Intensity_Level'] = data['Intensity'].apply(classify_intensity)

# Calories per minute
data['Cal_per_min'] = data['Calories'] / data['Duration'].replace(0, np.nan)  # avoid divide-by-zero

# Temperature Deviation from 37C
data['Temp_Deviation'] = data['Body_Temp'] - 37.0

# WHtR (Weight to Height Ratio)
data['WHtR'] = data['Weight'] / data['Height']

# VO2 proxy
data['VO2_proxy'] = data['Heart_Rate'] * data['Duration']

# Heart Rate x Body Temp interaction
data['HR_Temp'] = data['Heart_Rate'] * data['Body_Temp']

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

from sklearn.decomposition import PCA, FastICA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.manifold import TSNE
import umap.umap_ as umap

# Preprocessing
le = LabelEncoder()
data = data.copy()
data['Intensity_Level'] = le.fit_transform(data['Intensity_Level'])

X = data.drop(columns=['Calories'])
y = data['Calories']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Model evaluation utility
def evaluate_model(name, X_tr, X_te):
    model = RandomForestRegressor(random_state=42, n_jobs=-1)
    model.fit(X_tr, y_train)
    y_pred = model.predict(X_te)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    print(f"{name}: RMSE = {rmse:.2f}, R2 = {r2:.2f}")
    return rmse, r2

results = {}

# ------------------ Dimensionality Reduction ------------------

# 1. PCA (fast)
pca = PCA(n_components=6, random_state=42)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
results['PCA'] = evaluate_model("PCA", X_train_pca, X_test_pca)

# 2. ICA (moderate)
ica = FastICA(n_components=6, random_state=42)
X_train_ica = ica.fit_transform(X_train)
X_test_ica = ica.transform(X_test)
results['ICA'] = evaluate_model("ICA", X_train_ica, X_test_ica)

# 3. LDA (requires class labels)
y_class = pd.qcut(y, q=3, labels=False)
lda = LDA(n_components=1)
lda.fit(X_scaled, y_class)
X_train_lda = lda.transform(X_train)
X_test_lda = lda.transform(X_test)
results['LDA'] = evaluate_model("LDA", X_train_lda, X_test_lda)

# 4. Subsample for t-SNE & UMAP
sample_size = min(5000, len(X_scaled))
indices = np.random.choice(len(X_scaled), sample_size, replace=False)
X_sample = X_scaled[indices]
y_sample = y.iloc[indices]

print("Running t-SNE...")
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_sample)

print("Running UMAP...")
umap_model = umap.UMAP(n_components=2, random_state=42)
X_umap = umap_model.fit_transform(X_sample)

# ------------------ Visualization ------------------

def plot_2d(data_2d, title, labels, xlabel='Component 1', ylabel='Component 2'):
    plt.figure(figsize=(10, 7))
    scatter = plt.scatter(data_2d[:, 0], data_2d[:, 1], c=labels, cmap='viridis', s=10, alpha=0.7)
    cbar = plt.colorbar(scatter)
    cbar.set_label('Calories', fontsize=12)
    plt.title(title, fontsize=15)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True, linestyle='--', alpha=0.3)
    plt.tight_layout()
    plt.show()

# Visualize subsampled data only
plot_2d(PCA(n_components=2).fit_transform(X_sample), "PCA (2D) - Subsample", y_sample)
plot_2d(FastICA(n_components=2, random_state=42).fit_transform(X_sample), "ICA (2D) - Subsample", y_sample)
plot_2d(X_tsne, "t-SNE - Subsample", y_sample)
plot_2d(X_umap, "UMAP - Subsample", y_sample)

# Final Results
print("\nModel Performance Summary:")
for key, val in results.items():
    print(f"{key}: RMSE = {val[0]:.2f}, R2 = {val[1]:.2f}")
```

```python
import time
import numpy as np
import pandas as pd

from sklearn.linear_model import BayesianRidge, SGDRegressor, Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor, ExtraTreeRegressor
from sklearn.ensemble import AdaBoostRegressor, RandomForestRegressor, ExtraTreesRegressor, HistGradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.svm import LinearSVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

# Scale for linear models
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

models = {
    "BayesianRidge": (BayesianRidge(), X_train_scaled, X_test_scaled),
    "SGDRegressor": (SGDRegressor(max_iter=1000, tol=1e-3), X_train_scaled, X_test_scaled),
    "Lasso": (Lasso(alpha=0.1), X_train_scaled, X_test_scaled),
    "Ridge": (Ridge(alpha=1.0), X_train_scaled, X_test_scaled),
    "LinearSVR": (LinearSVR(max_iter=1000, random_state=42), X_train_scaled, X_test_scaled),
    
    "DecisionTreeRegressor": (DecisionTreeRegressor(random_state=42), X_train, X_test),
    "ExtraTreeRegressor": (ExtraTreeRegressor(random_state=42), X_train, X_test),
    "ExtraTreesRegressor": (ExtraTreesRegressor(n_estimators=50, random_state=42, n_jobs=-1), X_train, X_test),
    "RandomForestRegressor": (RandomForestRegressor(n_estimators=50, random_state=42, n_jobs=-1), X_train, X_test),
    "AdaBoostRegressor": (AdaBoostRegressor(n_estimators=50, random_state=42), X_train, X_test),
    "HistGradientBoosting": (HistGradientBoostingRegressor(max_iter=100, random_state=42), X_train, X_test),
    "XGBoost": (XGBRegressor(objective='reg:squarederror', n_estimators=50, random_state=42), X_train, X_test),
}

results = []

for name, (model, X_tr, X_te) in models.items():
    print(f"\n Training {name}...")
    start_time = time.time()
    model.fit(X_tr, y_train)
    duration = time.time() - start_time

    y_pred = model.predict(X_te)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    results.append({
        "Model": name,
        "Train Time (s)": round(duration, 2),
        "MAE": round(mae, 4),
        "RMSE": round(rmse, 4),
        "R2": round(r2, 4)
    })

results_df = pd.DataFrame(results).sort_values(by="R2", ascending=False)
print("\n All models trained. Summary:")
print(results_df)
```

```python

target_models = ["RandomForestRegressor", "ExtraTreesRegressor", "DecisionTreeRegressor"]

for model_name in target_models:
    model, X_tr, X_te = models[model_name] 
    
    y_pred_test = model.predict(X_te)
    y_pred_test = np.maximum(y_pred_test, 0)  
    submission_df = pd.DataFrame({
        "id": np.arange(len(X_te)),
        "Calories": y_pred_test
    })

    filename = f"submission_{model_name.lower()}.csv"
    submission_df.to_csv(filename, index=False)
    print(f" Submission saved: {filename}")
```
