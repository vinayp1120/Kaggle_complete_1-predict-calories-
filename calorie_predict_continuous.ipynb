```python
# Function to detect and remove outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Outliers in {column}:")
    print(outliers)
    
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df
```

```python
# Function to print correlated columns
def print_correlated_columns(correlation_matrix, column, threshold=0.7):
    correlated_columns = correlation_matrix[column][abs(correlation_matrix[column]) > threshold].index.tolist()
    correlated_columns.remove(column)
    print(f"Columns correlated with {column}: {correlated_columns}")

# Calculate correlation matrix
correlation_matrix = data.corr()

# Iterate over columns
for column in data.columns:
    if data[column].dtype in ['int64', 'float64']:  # Check if column is numerical
        print(f"\nColumn: {column}")
        print_correlated_columns(correlation_matrix, column)
        data = remove_outliers(data, column)

print("Data shape after removing outliers:", data.shape)


```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Function to plot distribution
def plot_distribution(data, column, title):
    plt.figure(figsize=(8,6))
    plt.hist(data[column].dropna(), bins=50, density=True, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title(title)
    plt.xlabel(column)
    plt.ylabel("Density")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Function to identify and transform distribution
def identify_distribution(data, column):
    col_data = data[column].dropna()

    # Perform normality test
    stat, p = stats.normaltest(col_data)
    if p < 0.05:
        print(f"{column} does not follow a normal distribution")
        
        # Try to identify the distribution
        distributions = ['norm', 'expon', 'uniform', 'poisson']
        results = []
        for dist in distributions:
            try:
                params = getattr(stats, dist).fit(col_data)
                d, p_value = stats.kstest(col_data, dist, args=params)
                results.append((dist, p_value))
            except Exception as e:
                continue
        
        if not results:
            print(f"No valid distribution fit for {column}. Skipping transformation.")
            return col_data

        # Get the best fit distribution
        best_dist = max(results, key=lambda x: x[1])  # best = highest p-value
        print(f"Best fit distribution for {column}: {best_dist[0]} (p={best_dist[1]:.4f})")

        # Apply transformation
        try:
            if best_dist[0] == 'expon':
                transformed_data = np.log(col_data + 1e-5)
            elif best_dist[0] == 'uniform':
                scaled = (col_data - col_data.min()) / (col_data.max() - col_data.min())
                transformed_data = stats.norm.ppf(scaled.clip(1e-5, 1 - 1e-5))
            elif best_dist[0] == 'poisson':
                transformed_data = np.sqrt(col_data)
            else:
                transformed_data = col_data  # already normal
        except Exception as e:
            print(f"Transformation failed for {column}: {e}")
            transformed_data = col_data

        return transformed_data
    else:
        print(f"{column} follows a normal distribution")
        return data[column]

# Iterate over numeric columns
for column in data.select_dtypes(include=['int64', 'float64']).columns:
    print(f"\nColumn: {column}")
    plot_distribution(data, column, f"Original Distribution of {column}")
    transformed_data = identify_distribution(data, column)
    data[column] = transformed_data
    plot_distribution(data, column, f"Transformed Distribution of {column}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Function to plot distribution
def plot_distribution(data, column, title):
    plt.figure(figsize=(8,6))
    plt.hist(data[column].dropna(), bins=50, density=True, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title(title)
    plt.xlabel(column)
    plt.ylabel("Density")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Function to normalize column using the best transformation
def normalize_column(data, column):
    col_data = data[column].dropna()

    # Skip if column has constant value
    if col_data.nunique() <= 1:
        print(f"{column} has constant value. Skipping.")
        return data[column]

    transformations = {
        'none': lambda x: x,
        'log': lambda x: np.log(x + 1e-5) if np.all(x > 0) else None,
        'sqrt': lambda x: np.sqrt(x) if np.all(x >= 0) else None,
        'boxcox': lambda x: stats.boxcox(x + 1e-5)[0] if np.all(x > 0) else None,
        'yeojohnson': lambda x: stats.yeojohnson(x)[0],
    }

    best_p = -np.inf
    best_trans = 'none'
    best_data = col_data

    for name, func in transformations.items():
        try:
            transformed = func(col_data)
            if transformed is not None and np.all(np.isfinite(transformed)):
                stat, p = stats.normaltest(transformed)
                if p > best_p:
                    best_p = p
                    best_trans = name
                    best_data = transformed
        except Exception as e:
            continue

    print(f"Best transformation for {column}: {best_trans} (p={best_p:.4f})")
    return pd.Series(best_data, index=col_data.index)

# Apply transformation to each numeric column
for column in data.select_dtypes(include=['int64', 'float64']).columns:
    print(f"\nColumn: {column}")
    plot_distribution(data, column, f"Original Distribution of {column}")
    transformed_column = normalize_column(data, column)
    data[column].loc[transformed_column.index] = transformed_column
    plot_distribution(data, column, f"Transformed Distribution of {column}")
```

```python
#feature engineering
'''
1.BMI = Weight / (Height/100)²
Max_HR = 220 - Age
2.Intensity = Heart_Rat4.e / Max_HR
if Intensity < 0.6: Low
elif Intensity < 0.75: Moderate
else: High
3.Cal_per_min = Calories / Duration
4.Temp_Deviation = Body_Temp - 37.0
5.WHtR = Weight / Height
6.VO2_proxy = Heart_Rate * Duration
7.HR_Temp = Heart_Rate * Body_Temp
'''
import pandas as pd
import numpy as np

# Assuming data is your original DataFrame
# Example: data = pd.read_csv("your_data.csv")

# BMI
data['BMI'] = data['Weight'] / ((data['Height'] / 100) ** 2)

# Max Heart Rate (220 - age)
data['Max_Heart_Rate'] = 220 - data['Age']

# Intensity of workout
data['Intensity'] = data['Heart_Rate'] / data['Max_Heart_Rate']

# Intensity Category
def classify_intensity(i):
    if i < 0.6:
        return 'Low'
    elif i < 0.75:
        return 'Moderate'
    else:
        return 'High'

data['Intensity_Level'] = data['Intensity'].apply(classify_intensity)

# Calories per minute
data['Cal_per_min'] = data['Calories'] / data['Duration'].replace(0, np.nan)  # avoid divide-by-zero

# Temperature Deviation from 37°C
data['Temp_Deviation'] = data['Body_Temp'] - 37.0

# WHtR (Weight to Height Ratio)
data['WHtR'] = data['Weight'] / data['Height']

# VO2 proxy
data['VO2_proxy'] = data['Heart_Rate'] * data['Duration']

# Heart Rate x Body Temp interaction
data['HR_Temp'] = data['Heart_Rate'] * data['Body_Temp']

```
